## Which movie made the highest profit? Who were its producer and director? Identify the actors in that film.

```python
# Getting the profit
data['profit'] = data['revenue'] - data['budget']
```

```python
# Crew column to ecxtract/mine director and producer

print(data['crew'][0])
print(type(data['crew'][0]))
```

```python
# Function to extract the Director Name
def getDirectorName(temp):
    try:
        # Converting it to list
        temp_list = ast.literal_eval(temp)
        # Extract the director name
        d_name =''
        for d in temp_list:
            if d['job'] == 'Director':
                d_name = d['name']

        return d_name
    except:
        return ''
```

```python
data['director'] = data['crew'].apply(getDirectorName)
```

```python
def getProducers(x):
    try:
        # convert string to list - literal_eval
        var1_list = ast.literal_eval(x)

        # Extracting the producer name
        p_list = []
        for d in var1_list:
            if d['job'] =='Producer':
                p_name = d['name']
                p_list.append(p_name)

        return p_list
    except:
        return []
```

```python
data['producers'] = data['crew'].apply(getProducers)
```

```python
def getActors(x):
    try:
        # convert string to list - literal_eval
        var1_list = ast.literal_eval(x)

        # Extracting the producer name
        a_list = []
        for d in var1_list:
            a_name = d['name']
            a_list.append(a_name)

        return a_list[:4]
    except:
        return []
```

```python
data['actors'] = data['cast'].apply(getActors)
```

```python
# get the maximum profit
highest_profit  = data['profit'].max()

temp_data = data[data['profit'] == highest_profit]

temp_data[['title', 'profit', 'director', 'producers', 'actors']]
```

# Find out the unique genres of movies in this dataset.

```python
def getGenres(x):
    try:
        # Convert it to list
        x_list = ast.literal_eval(x)

        # Empty list to store genre
        g_list = []
        for d in x_list:
            g_list.append(d['name'])
        
        return g_list
    except:
        return []

# Store the list of genres into a new column
data['genres_list'] = data['genres'].apply(getGenres)
```
